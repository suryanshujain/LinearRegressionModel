```python
x_train=np.array(699)
y_train=np.array(699)
```

```python
print(train.keys())
```

```python
x_train=train.x[:]
y_train=train.y[:]
```

```python
def cost(x,y,w,b):
  m=train.shape[0]
  cost=0
  for i in range(m):
    y_pred=w*x[i]+b
    cost=cost+((y_pred-y[i])**2)
  net_cost=cost/(2*m)
  return net_cost
```

```python
def gradient(x,y,w,b):
  m = x.shape[0]
  dj_dw = 0
  dj_db = 0
  for i in range(m):
    y_pred=w*x[i]+b
    dj__dw=(y_pred - y[i])*x[i]
    dj__db=y_pred-y[i]
    dj_dw+=dj__dw
    dj_db+=dj__db
  dj_dw=dj_dw/m
  dj_db=dj_db/m

  return dj_dw, dj_db
```

```python
def descent(x,y,w0,b0,a,itr,cost,gradient):
  pw=[]
  pb=[]
  w=w0
  b=b0
  for i in range(itr):
    dj_dw,dj_db=gradient(x,y,w,b)
    w-=a*(dj_dw)
    b-=a*(dj_db)
    print(cost(x,y,w,b))
    pw.append(w)
    pb.append(b)
  return w,b,pw,pb
```

```python
w0=0
b0=0
itr=10000
a=1e-6
w_net,b_net,pw,pb= descent(x_train,y_train,w0,b0,a,itr,cost,gradient)
```

```python
print(w_net,b_net)
```

```python
path1="/content/drive/MyDrive/Dataset/test.csv"
test=pd.read_csv(path1)
```

```python
x_test=test.x[:]
y_test=test.y[:]
```

```python
n=x_test.shape[0]
print(n)
```

```python
err=0
for i in range(n):
  y_pred=w_net*x_test[i]+b_net
  b=(y_pred-y_test[i])**2
  err+=b
err/=n
rms=(err)**(1/2)
print(rms)
```
